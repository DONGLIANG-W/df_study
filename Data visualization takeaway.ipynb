### Data visualization

basic plotting with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([5,6],[5,7],'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot y vs x figure')
plt.show()
```

first create a dataframe, and draw the dataframe with plt

```python
import pandas as pd
df = pd.DataFrame({'year':[1980,1981,1982,1983,1984], 'India':[8880,8670,8147,7338,5704], 'China':[5123,6682,3308,1863,1527]})
df['year'] = df['year'].astype('int32')
df.dtypes
```

```python
df.set_index('year', inplace=True)
df.head()
```

simply using df.plot(), the pandas matplotlib will plot the figure

```python
df.plot(kind='line')
```

```python
df['India'].plot(kind='hist')
```

df.loc is used to slice the dataframe with fixed index

```python
df.loc[1981]
```

```python
df.loc[1981]['India']
```

sort the df by value

```python
df.sort_values('China', ascending=True, axis=0, inplace=True)
df.head()
```

```python
# df.reset_index(inplace=True)
df.sort_values('year',inplace=True)
df.loc[:].astype('int64')
```

```python
df.plot(kind='area')
plt.xlabel("Year")
plt.ylabel("immigration population")
plt.title(" immigration population over year")
```

```python
df.dtypes
```

pandas.hist() or df.plot(kind='hist') are able to plot the histogram of the data

```python
df['India'].hist()
```

```python
df.plot(kind='hist')
```

one point that improve is that the bins are not aligned with the tick marks on the horizontal axis. this can make the histogram hard to read, we can fix this by calling numpy histogram function, what this function is going to do is it partition the spread of the data in column into 10 bins of equal width, compute the number of datapoints that fall in each bin, and then return this frequency of each bin which we are calling count here and bin edges which we're calling bin_edges. 

```python
count, bin_edges = np.histogram(df)
print('count is {} and edges is {}'.format(count, bin_edges))
```

```python
df.plot(kind='hist', xticks=bin_edges)
plt.xlabel('number of immigrants')
plt.ylabel('number of countries')
plt.title('Histogram of immigration over years')
```
